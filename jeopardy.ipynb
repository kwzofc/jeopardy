```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
# Read CSV File And Rename Columns
all_data = pd.read_csv('jeopardy.csv').dropna(how='all').rename(columns={
    'Show Number': 'show_number',
    ' Air Date': 'air_date',
    ' Round': 'round',
    ' Category': 'category',
    ' Value': 'value',
    ' Question': 'question',
    ' Answer': 'answer'
})
```

```python
all_data
```

```python
# Search Function
def search(stringset):
    result = all_data
    for string in stringset:
        result = result[result.question.str.contains(string, case=False, regex=True)]
    return result
```

```python
search(['King', 'England'])
```

```python
#Add new column of value
all_data['float_value'] = all_data.apply(lambda row: row.value[1:].replace(',', '') if row.value != 'None' else 0.0, axis=1)
```

```python
all_data
```

```python
# Change data type to float
all_data['float_value'] = pd.to_numeric(all_data['float_value'])
```

```python
all_data
```

```python
# Average Value of Questions That Contain The Word "King"
average_values_of_king_word = search(['King']).float_value.mean()
print('Average question values is ${}'.format(round(average_values_of_king_word, 2)))
```

```python
# Function that returns the count of the unique answers to all of the questions in a dataset.
def unique_answer(search):
    question = search
    result = question.answer.nunique()
    return result
```

```python
# Filtering the entire dataset to only questions containing the word "King" and count all of the unique answers in dataset.
king_word_unique_answer = unique_answer(search(['King']))
print(f'Total answers: {king_word_unique_answer}')
```
